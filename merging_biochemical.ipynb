```python
import pandas as pd
import numpy as np
```

```python
data_spectrum_gt = pd.read_csv("sorghum_2022_whole.csv")
```

```python
data_bio = pd.read_csv("biochemical.csv")
```

```python
data_bio.head(50)
```

```python
# Removing NaN columns in data_bio
data_bio = data_bio.dropna(axis=1)
```

```python
data_bio.columns
```

```python
# Removing unnecessary columns in data_bio
data_bio.drop(['Kind Of Sample', 'Cust ID', 'Name', 'Company', 'Address 1', 'City', 'State', 'Zip', 'Lab No', 'Grower', 'Field ID', 'Plant Type', 'Plant Stage'], axis=1, inplace=True)
```

```python
data_bio.head()
```

```python
data_bio["Sample ID"]
```

```python
Genotype = []
Rep = []
for i in data_bio['Sample ID'] :
    str(i)
    a, b = i.split('-')
    Genotype.append(a)
    Rep.append(b)  
    
```

```python
cols = [Genotype, Rep]
places = [2,3]
names = ['Genotype', 'Rep']
for i, j, k in zip(cols, places, names):
    data_bio.insert(j , k , i)
```

```python
data_bio['Genotype'] = data_bio['Genotype'].astype('int64')
```

```python
data_bio['Rep'] = data_bio['Rep'].astype('int64')
```

```python
data_spectrum_gt['Rep'] = data_spectrum_gt['Rep'].astype('int64')
```

```python
data_spectrum_gt['Rep'].value_counts(dropna=False)
```

```python
data_bio['Rep'].value_counts(dropna=False)
```

```python
data= pd.merge(data_bio, data_spectrum_gt, how='outer')
```

```python
data.head(50)
```

```python
data.columns[0:30]
```

```python
data.rename(columns={'% P': 'P', '% K' : 'K', '% S': 'S', '% Ca': 'Ca', '% Mg': 'Mg', 'ppm Zn' : 'Zn', 'ppm Fe': 'Fe', 
                     'ppm Mn': 'Mn', 'ppm Cu': 'Cu', 'ppm B': 'B', 'ppm Mo': 'Mo'}, inplace=True)
```

```python
data.to_csv('sorghum_2022_whole.csv', index=False)
```
